```python
from sklearn.datasets import make_classification

X, y = make_classification(n_samples=100, n_features=2, n_classes = 2, n_clusters_per_class=1, n_informative = 1, n_redundant=0,
                           random_state=41, hypercube=False, class_sep = 10)

```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(10,10))
plt.scatter(X[:,0], X[:,1], c=y)
plt.show()
```

```python
import numpy as np

def perceptron(X,y):
  X = np.insert(X,0,1,axis=1)
  weights = np.ones(X.shape[1])
  lr = 0.1

  for i in range(1000):
    j = np.random.randint(0,100)
    y_hat = step(np.dot(weights,X[j]))
    weights = weights + lr*(y[j] - y_hat)*(X[j])

  return weights[0], weights[1:]
```

```python
def step(z):
  return int(z>0)
```

```python
bias, weights = perceptron(X,y)
```

```python
m = -(weights[0]/weights[1])
b = -(bias/weights[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```

```python
plt.figure(figsize=(10,10))
plt.scatter(X[:,0], X[:,1], c=y)
plt.plot(x_input,y_input)
plt.show()
```
